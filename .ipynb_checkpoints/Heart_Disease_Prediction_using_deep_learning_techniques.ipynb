# Importing pakages and libraries

```python
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import sklearn
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
from sklearn.preprocessing import MinMaxScaler

from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix
```

# Importing dataset and merging

```python
# loading data into pandas data frame

heart_data_1 = pd.read_csv("/content/heart_dataset_RafaelGranza_1026.csv")
heart_data_1.head(10)
```

```python
heart_data_2 = pd.read_csv("/content/heart_dataset_bharath0609_304.csv")
heart_data_2.head(10)
```

```python
heart_data_3 = pd.read_csv("/content/heart_dataset_jocelyndumlao_1000.csv")
heart_data_3.head(10)
```

```python
heart_data_4 = pd.read_csv("/content/heart_dataset_rishidamarla_271.csv")
heart_data_4.head(10)
```

```python
heart_data_1.shape
```

```python
heart_data_2.shape
```

```python
heart_data_3.shape
```

```python
heart_data_4.shape
```

```python
heart_data = pd.concat([heart_data_1, heart_data_2, heart_data_3, heart_data_4])
heart_data.shape
```

```python
heart_data.head()
```

# Splitting into features and target

```python
target_column = "target"
numerical_column = heart_data.columns.drop(target_column)
output_rows = heart_data[target_column]
heart_data.drop(target_column,axis=1,inplace=True)
```

```python
# # prompt: split features and target into x and y, y is last columns of the dataset

# X = heart_data.iloc[:, :-1]
# X.head()
```

```python
# Min-Max scaling Implementation in python

scaler = MinMaxScaler()
scaler.fit(heart_data)
t_df = scaler.transform(heart_data)
```

```python
# y = heart_data.iloc[:, -1]
# y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(t_df, output_rows, test_size=0.25, random_state=42)
```

```python
print('X_train:',np.shape(X_train))
print('y_train:',np.shape(y_train))
print('X_test:',np.shape(X_test))
print('y_test:',np.shape(y_test))
```

# Applying deep learning


```python
# define the keras model

model = Sequential()
model.add(Dense(units=16, activation='relu', input_shape=(13,)))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# compile the keras model

adam = keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss='binary_crossentropy', optimizer=adam, metrics=["accuracy"])
```

```python
model.fit(X_train, y_train, epochs=150, batch_size = 50)
```

# Accuracy metrices

```python
predicted = model.predict(X_test)
```

```python
predicted = tf.squeeze(predicted)
predicted = np.array([1 if x >= 0.5 else 0 for x in predicted])
actual = np.array(y_test)
conf_mat = confusion_matrix(actual, predicted)
displ = ConfusionMatrixDisplay(confusion_matrix=conf_mat)
displ.plot()
```

```python
loss_and_metrics = model.evaluate(X_test, y_test)
print(loss_and_metrics)
print('Loss = ',loss_and_metrics[0])
print('Accuracy = ',loss_and_metrics[1])
```

```python
print('Model Precision score: ', precision_score(y_test, predicted))
```

```python
# check results
print(metrics.classification_report(y_test, predicted))
```
